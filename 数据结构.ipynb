```python
'''
字典和集合的性能
'''
#列表查询如下：时间复杂度为 O(n)
def find_product_price(products, product_id):
    for id,price in products:
        if id == product_id:
            return price
    return None

products = [
    (111111,100),
    (222222,30),
    (555555,150)
]

print('The price of product of 222222 is {}'.format(find_product_price(products,222222)))
```

```python
#使用字典查询，时间复杂度 O(1)
def find_product_price(products, product_id):
    if product_id in products:
        return products[product_id]
    return None

products = {
    111111: 100,
    222222: 30,
    555555: 150
}
print('The price of product of 222222 is {}'.format(find_product_price(products,222222)))
```

```python
#记录由多少中不同价格
#列表
def find_product_price(products):
    unique_price_list = []
    for _,price in products:
        if price not in unique_price_list:
            unique_price_list.append(price)
    return len(unique_price_list)

products = [
    (111111,100),
    (222222,30),
    (555555,150),
     (333,40),
     (555,150),
     (9999,30)
]
print('number of unique price is {}'.format(find_product_price(products)))
```

```python
#使用集合
def find_product_price(products):
    unique_price_set = set()
    for _,price in products:
        unique_price_set.add(price)
    return len(unique_price_set)


products = [
    (111111,100),
    (222222,30),
    (555555,150),
     (333,40),
     (555,150),
     (9999,30)
]
print('number of unique price is {}'.format(find_product_price(products)))
```

```python
'''
字典和集合的工作原理：
字典和集合的内部结构都是一张哈希表，对于字典而言，这张表存储了哈希值(hash)、键和值这3个元素。
对集合而言，区别就是哈希表内没有键和值的配对，只有单一的元素。

插入操作：
每次向字典或集合插入一个元素时，python会首先计算键的哈希值(hash(key))，再和 mask = PyDicMinSize -1 做与操作，计算这个元素应该插入哈希表的位置 index = hash(key) & mask 。 
如果哈希表中此位置是空的，那么这个元素就插入其中；如果此位置被占用，Python便会比较两个元素的哈希值和键是否相等：如果相等，则表明这个元素已经存在，
如果值不同，则更新值；若两者中有一个不相等，这种情况我们通常称为哈希冲突（hash collision）,意思是两个元素的键不相等，但是哈希值相等，这种情况，Python便会继续寻找表中空余位置，直到找到位置为止。

查找操作：
和前面的插入操作类似，Python会根据哈希值，找到其对应的处于的位置；然后，比较哈希表这个位置中元素的哈希值和键，与需要查找的元素是否相等。如果相等，则直接返回；
如果不等，则继续查找，直到找到空位或者抛出异常为止。

删除操作：
对于删除操作，Python会暂时对这个位置的元素，赋予一个特殊的值，等到重新调整哈希表的大小时，再将其删除。

哈希冲突的发生，往往会降低字典和集合操作的速度。因此为了保证其高效性，字典和集合内的哈希表，通常会保证至少留有1/3的剩余空间。随着元素的不停插入，
当剩余空间小于1/3时，Python会重新获取更大的内存空间，扩充哈希表。不过这种情况表内所有元素位置都会被重新排放
虽然哈希冲突和哈希表大小的调整，都会导致速度减缓，但是这种情况发生的次数极少，所以平均情况下，这仍能保证插入、查找和删除的时间复杂度为O(1)

'''
```
