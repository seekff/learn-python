错误和异常

错误包含两种，一种是语法错误，一种是异常


异常是语法正确，也可以被执行，但是执行过程中遇到了错误，抛出异常。

如何处理异常：

通常使用 try 和 except 来解决，还可以增加finally

处理异常时，如果必须执行的语句，可以放在 finnally block 中。

异常处理，通常用在不确定某段代码能否成功执行，也无法轻易判断的情况下，比如数据库连接、读取扥扥

```python


try:
  s = input('please entrer tow numbers :')
  num1 = int(s.split(',')[0])
  num2 = int(s.split(',')[1])
  print(num1+num2)
except ValueError as err:
  print('value error: {}'.format(err))
except IndexError as err:
  print('index error: {}'.format(err))
except:
  print('other error')
finally:
  print('finally  block')
```

```python
try:
  s = input('please entrer tow numbers :')
  num1 = int(s.split(',')[0])
  num2 = int(s.split(',')[1])
  print(num1+num2)
except (ValueError,IndexError) as err:
  print('value error: {}'.format(err))
except Exception as err:
  print('other error')
finally:
  print('finally  block')
```
