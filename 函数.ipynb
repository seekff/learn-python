**函数的变量作用域**

局部变量： 变量在函数内部定义，只在函数内部有效，一旦函数执行完成，局部变量就会被回收，无法访问。

例如下面函数内部定义了 file 这个变量，只在read_text_from_file 这个函数有效，在函数外部无法访问：

    def read_text_from_file(path):

      with open(path) as file:

        ...


全局变量：定义在整个文件层次上的，可以在文件内任何地方被访问，不过不能在函数内部随意改变全局变量的值。一定要函数内修改全局变量的值，则需要使用 global 关键字,这个关键字告诉python解析器，函数内部的这个变量是之前定义的全局变量，并不是全新的全局变量，也不是局部变量。

例如：MIN_VALUE和MAX_VALUE就是全局变量

    MIN_VALUE = 1
    MAX_VALUE = 100

    def validation_cherk(value):

      if value < MIN_VALUE or MIN_VALUE > MAX_VALUE:

        raise Excption('validation check fails')



如果遇到函数内部的局部变量和全局变量同名，那么在函数内部，局部变量会覆盖全局变量




```python
MIN_VALUE = 1
MAX_VALUE = 100

def validation_cherk_g(value):

  global MIN_VALUE

  if value <5:
    MIN_VALUE += value
    print('函数内部 MIN_VALUE = {}'.format(MIN_VALUE))


def validation_cherk_j():

  MAX_VALUE = 20

  print('函数内部 MAX_VALUE = {}'.format(MAX_VALUE))

validation_cherk_g(3)

validation_cherk_j()

print('全局 MIN_VALUE = {}'.format(MIN_VALUE))
print('全局 MAX_VALUE = {}'.format(MAX_VALUE))
```

**函数的嵌套**

1、保证内部函数的隐私，内部函数只能被外部函数所调用和访问，不会暴露在全局作用域

2、合理的使用函数嵌套，能够提高程序的运行效率。

对于嵌套函数来说，内部函数可以访问外部函数定义的变量，但是无法修改，若要修改，必须加上nonlocal 这个关键字




```python
def factor(input):
  if not isinstance(input, int):
    raise Exception('input must be integer.')
  if input < 0:
    raise Exception('input must be greater or equal to 0')

  def inner_factor(input):
    if input <= 1:
      return 1
    return input * inner_factor(input-1)
  return inner_factor(input)

print(factor(5))
```

```python
def outer():
  x = 'local'
  def inner():
    nonlocal x
    x = 'nonlocal'
    print('inner:',x)
  inner()
  print('outer:',x)

outer()
```

**闭包**

与嵌套函数相似，不同的是，外部函数返回的是一个函数，而不是一个具体的值。返回的函数通常赋于一个变量，这个变量在后面继续被执行调用。


```python
def nth_power(exponent):
  def exponent_of(base):
    return base ** exponent
  return exponent_of

square = nth_power(2) #计算一个数的平方
cube = nth_power(3) #计算一个数的立方

print(square)
print(cube)

print(square(2))
print(cube(5))
print(cube(2))
```

**匿名函数**

格式： lambda argument1, argument2, argument3, ... argumentM : expression

匿名函数的关键字是 **lambda** ，之后一系列的参数，然后冒号隔开，最后则是由这些参数组成的表达式。

    square = lambda x: x**2

    square(3)

等价于

    def square(x):

      return x**2

    square(3)


与其他常规函数的区别：

**第一，lambda 是一个表达式(expression)，并不是一个语句（statement）**

所谓表达式，就是用一些列“公式”去表达一个东西，比如 x+2、x**2等到；

所谓的语句，则是一定完成了某些功能，比如赋值语句 x=1 完成了赋值，print语句 print(x) 完成了打印，条件语句 if  x< 0 ： 完成了选择功能等到

**第二，lambda 的主体是只有一行的简单表达式，并不能扩展成一个多行的代码块。**



```python
square = lambda x: x**2
print(square(3))

l = [(lambda x: x*x)(x) for x in range(10)]
print(l)

l = [(1,20),(3,0),(9,10),(2,-1)]
l.sort(key=lambda x: x[1]) #按照列表中元组的第二个元素排序
print(l)
```

**函数式编程**

函数式编程，是指代码中每一块都是不可变的，都是由纯函数的形式组成。这里的纯函数，是指函数本身相互独立、互不影响，对于相同的输入，总会由相同的输出，没有任何副作用。

例如：下面就不是函数式编程，多次调用，每次得到的结果都不一样

    def nultipy_2(l):

      for index in range(0. len(l)):

        l[index] *= 2

      return l
    
函数式编程：

    def nultipy_2_pure(l):

      new_list = []

      for item in l :

        new_list.append(item * 2)

      return new_list



函数式编程优点：主要在于其纯函数和不可变的特性使程序更加健壮，易于调试和测试；

缺点：主要在于限制多，难写。


**python主要提供这么几个函数： map()、filter()和reduce()，通常集合匿名函数 lambda 一起使用。**

函数**map(function, iterable) **的第一个参数是函数对象，第二个参数是一个可以遍历的集合，它表示对iterable的每一个元素，都运用function这个函数。


函数**filter(function, iterable)**函数，它和 map 函数类似，functon同样表示一个函数对象。filter() 函数表示对 iterable 中的每一个元素，都是用function判断，并返回True或者False，最后将返回True的元素组成一个新的遍历的集合


函数 **reduce(function, iterable)**，通常用来对一个集合做一些累计的操作。functon 同样是一个函数对象，规定它有两个参数，表示对iterable中的每个元素以及上一次调用后的结果，运用function进行计算，所以最后返回的是一个单独的数值。(**注意：reduce 函数在 Python 3 中不再是内置函数，它需要从 functools 模块中导入才能使用。**)


```python

square = map(lambda x: x**2, [1,2,3,4,5])

square

print(list(square))
```

```python
l = [1,2,3,4,5]
new_list = map(lambda x: x*2, l)

print(list(new_list))
```

```python
l = [1,2,3,4,5]
new_list = filter(lambda x: x%2==0, l)

print(list(new_list))
```

```python
from functools import reduce

l = [1,2,3,4,5]
product = reduce(lambda x,y : x*y, l)

print(product)
```
