**循环语句：**
for循环和while循环

数据结构只要有可迭代的（iterable），比如列表、集合等，那么都可以用以下方式遍历：

for item in <iterable>:

      ...


**注意**：字典本身只有键是可迭代的，如果遍历它的值或者键值对，需要其内置函数
values（）和item()实现，values()返回字典的值集合，items（）返回键值对的集合。

**while循环表示当condition满足时，一直重复循环内部的操作，直到condition不在满足，就跳出循环**

while conticon:

      ......


```python
d = {'name': 'jack', 'age': 30, 'gender': 'male'}

for  k in d:
  print(k)
print('-------------------------------')
for k,v in d.items():
  print('key:{},value:{}'.format(k,v))

print('************************')
for v in d.values():
  print(v)
```

```python
#使用range()函数，拿到索引，在通过索引在遍历集合中的元素。
l = [1,2,3,4,5,6,7,8]
for index in range(len(l)):
    if index < 5:
      print(l[index])

print('---------------')
for index in range(0,len(l)): #range(2,len(l)) 从第三个索引开始
    if index < 5:
      print(l[index])
```

```python
#内置enumerate()函数可以同时获取索引和元素
l = [6,2,3,4,5]
for index, item in enumerate(l):
  print('indxe:{},item:{}'.format(index,item))
```

```python
l = [6,2,3,4,5]
index = 0
while index < len(l):
  print(l[index])
  index += 1
```

**条件与循环的复用**

expression1 if condition else expression2 for item in iterable

这个表达式分解等同于下面成嵌套：

    for item in iterable:

      if condition:

        expression1

      else:

        expression2


如果没有 else 则需要写成如下：

expression for item in iterable if condition

分解嵌套：

    for item in iterable:

      if condition:

        expression


例子：y = 2*|x| +5  

y = [ value * 2 +5 if x >0  else -value * 2 + 5  for  value  in  x ]




```python
#将一个字符串，按照逗号分隔单词，去掉首位空字符，并过滤长度小于3的单词，最后返回单词列表

text = ' I, like, english , and, chinese '
words = [word.strip() for word in text.split(',') if len(word.strip()) > 3 ]
print(words)
```
