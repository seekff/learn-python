**1. Google Python 风格指南概述**

Google 的 Python 代码风格指南主要包括以下几个部分：

代码风格（Code Style）

命名规范（Naming Conventions）

文档字符串（Docstrings）

错误处理（Error Handling）

导入模块（Imports）

注释规范（Comments）

测试（Testing）

性能优化（Performance Optimizations）


**2. 代码风格（Code Style）**

Google 遵循 PEP 8 作为代码风格标准，同时在某些方面做了增强，确保代码更易读、易维护。

2.1 缩进
✅ 使用
4 个空格进行缩进（而不是 Tab）

    def my_function():
        x = 10
        print(x)

❌ 不要使用 Tab 进行缩进

    def my_function():
    →→x = 10  # (❌ 不要使用 Tab)
    →→print(x)

2.2 每行字符数

✅ 每行最多 80 个字符

    def calculate_area(width, height):
        return width * height
❌ 避免超长代码

    def calculate_area_of_rectangle_with_given_width_and_height(width, height):  # (❌ 超长)
        return width * height
长行代码可以换行（推荐使用括号或 \）

    my_list = [
        "apple", "banana", "cherry",
        "date", "fig", "grape"
    ]
2.3 变量赋值

✅ 在等号 = 两侧加空格

    x = 10
    y = x + 5

❌ 不要省略空格

    x=10  # (❌)
    y=x+5  # (❌)

**3. 命名规范（Naming Conventions）**

Google 遵循 PEP 8 的命名规范，并做了一些增强：

名称类型	命名方式	示例

变量	小写 + 下划线	user_name

函数	小写 + 下划线	get_user_info()

类	首字母大写的驼峰命名	UserProfile

常量	全部大写 + 下划线	MAX_CONNECTIONS

私有变量	_前缀	_secret_key

不要使用 l、O、I 作为变量名，因为它们容易与 1 和 0 混淆

    l = 10  # (❌)
    O = "object"  # (❌)

**4. 文档字符串（Docstrings）**

Google 推荐使用 三重双引号 """ 进行函数、类、模块的文档描述，格式如下：

4.1 函数文档

    def add_numbers(a: int, b: int) -> int:
        """计算两个数的和。

        Args:
            a (int): 第一个数字
            b (int): 第二个数字

        Returns:
            int: 两个数字的和
        """
        return a + b
适用于：

函数/类/模块的说明

参数、返回值描述


**5. 错误处理（Error Handling）**

Google 推荐使用 try-except 进行异常处理，并尽量捕获特定异常。

✅ 正确示例

    try:
        result = 10 / 0
    except ZeroDivisionError as e:
        print(f"发生错误: {e}")

❌ 不要使用 except: 捕获所有异常

    try:
        result = 10 / 0
    except:  # (❌ 这样会隐藏所有错误)
        print("发生错误")

推荐：明确指定异常类型，如 ValueError、IOError

**6. 导入模块（Imports）**

Google 推荐：

每个 import 语句占一行

标准库、第三方库、自定义库分组

✅ 正确示例

    import os
    import sys

    import requests  # 第三方库

    from my_project import utils  # 自定义模块
❌ 不要在一行中导入多个模块

    import os, sys  # (❌)

**7. 注释规范（Comments）**

Google 推荐使用 完整的英文句子进行注释，并在 句号后加空格。

✅ 正确示例

    # 计算圆的面积
    def calculate_area(radius):
        return 3.14 * radius ** 2
❌ 错误示例

    #计算圆的面积
    def calculate_area(radius):
        return 3.14 * radius ** 2
注释应清晰描述代码逻辑，而不是重复代码

    x = x + 1  # (❌ 不要写 "增加 x 的值")

**8. 单元测试（Testing）**

Google 推荐使用 unittest 进行测试，确保代码可靠性。

8.1 创建单元测试

    import unittest

    def add(a, b):
        return a + b

    class TestMathOperations(unittest.TestCase):
        def test_add(self):
            self.assertEqual(add(2, 3), 5)

    if __name__ == "__main__":
        unittest.main()

推荐：每个测试方法名以 test_ 开头

**9. 代码性能优化**

Google 提倡：

使用 lru_cache 进行缓存

使用 set 进行去重

使用 list comprehension 替代 for 循环

✅ 正确示例

    from functools import lru_cache

    @lru_cache(maxsize=100)
    def fetch_data():
        return "Data"
❌ 避免低效代码

    # 使用 list 进行重复查找（低效）
    if item in my_list:
        my_list.remove(item)

**10. 总结**

Google 的 Python 惯用法强调：

✅ 代码简洁：避免冗余代码，提高可读性。

✅ 命名规范：符合 PEP 8，使用一致的变量、函数、类命名。

✅ 文档友好：使用 docstrings 提供清晰的 API 说明。

✅ 异常处理：捕获特定异常，避免 except: 随意捕获所有错误。

✅ 模块管理：按顺序导入标准库、第三方库、自定义模块。

✅ 注释清晰：注释应简明扼要，解释代码逻辑而非重复代码。
